```python
#pandas for handling data efficiently.pd is used instead of pandas
import pandas as pd
```

```python
#function read_csv is used from pandas 
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#prints the loaded data
dataset
```

```python
#input split
independent=dataset[["YearsExperience"]]
independent
```

```python
#input split
dependent=dataset[["Salary"]]
dependent
```

```python
#to split training and testing data 
from sklearn.model_selection import train_test_split
#function with parameters
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Linear regression to predict the targetted output. It is imported from sklearn
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
#fit trains the linear regression model
regressor.fit(X_train,y_train)
```

```python
#to calculate the slope
weight=regressor.coef_
weight
```

```python
#to calculate the initial value
bias=regressor.intercept_
bias
```

```python
#Checking the model
y_pred=regressor.predict(X_test)
```

```python
# R2 value is calculated by comparing actual output with predicted output
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#deployment phase
import pickle
#file is created with proper extension of pickle
filename="finalized_model_linear.sav"
```

```python
#data loading process takes place
pickle.dump(regressor,open(filename,"wb"))
```

```python
#loads the trained linear regression model to predict the output of given data
loaded_model=pickle.load(open("finalized_model_linear.sav","rb"))
result=loaded_model.predict([[15]])
```
