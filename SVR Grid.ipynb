```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.31,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

param_grid = {
    'kernel': ['rbf', 'poly', 'sigmoid', 'linear'],
    'C': [10, 100, 1000, 2000, 3000],
    'gamma': ['scale']  # Removed 'auto' to avoid deprecation warnings
}

grid = GridSearchCV(SVR(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='neg_mean_squared_error')
grid.fit(X_train,y_train)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
